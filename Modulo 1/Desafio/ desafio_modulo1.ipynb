#Análise de dados

##Aplicando método do cotovelo

```python
from sklearn.cluster import KMeans
```

```python
def calcular_wcss(dados_clientes):
  wcss = []
  for k in range(1,11):
    kmeans = KMeans(n_clusters = k, random_state=0)
    kmeans.fit(X=dados_cliente)
    wcss.append(kmeans.inertia_)
  return wcss
```

```python
# Criando uma nova base para análise de cluster
dados_cliente = base_completa[['peso', 'colesterol']]
dados_cliente.head()
```

```python
wcss_clientes = calcular_wcss(dados_cliente)
```

```python
type(wcss_clientes)
```

```python
#Visualizando dados obtidos no WCSS

for i in range(len(wcss_clientes)):
  print(f'O cluster {i} possui valor de WCSS de: {wcss_clientes[i]}')
```

##Visualizando gráfico de cotovelo

```python
import plotly.express as px
import plotly.offline as py
import plotly.graph_objects as go
```

```python
grafico_wcss = px.line(x= range(1,11), y=wcss_clientes)

fig = go.Figure(grafico_wcss)

fig.update_layout(title='Calculando WCSS', xaxis_title='número de cluster', yaxis_title='valor do WCSS', template='plotly_white')
fig.show()
```

```python
kmeans_clientes = KMeans(n_clusters=4, random_state=0)
base_completa['cluster'] = kmeans_clientes.fit_predict(dados_cliente)
```

```python
base_completa
```

##Calcular centroides de cada cluster

```python
centroides_clusters = kmeans_clientes.cluster_centers_
centroides_clusters
```

```python
dados_cliente
```

```python
grafico = px.scatter(x = base_completa['peso'], y = base_completa['colesterol'], color = base_completa['cluster'])

grafico_centroide = px.scatter(x = centroides_clusters[:,0], y = centroides_clusters[:,1], size = [7,7,7,7])
grafico_final_petala = go.Figure(data = grafico.data + grafico_centroide.data)
grafico_final_petala.show()
```

##Visualizando e alterando layout do gráfico

```python
grafico = px.scatter(x = base_completa['colesterol'], y = base_completa['peso'], color = base_completa['cluster'])

fig = go.Figure(grafico)

fig.update_layout(title='Análise de cluster', xaxis_title= 'Colesterol', yaxis_title= 'Peso')
fig.show()
```

##Criando função para gerar gráficos

```python
def plotar_graficos(x,y, xlabel, y_label):
  grafico = px.scatter(x = x, y = y, color = base_completa['cluster'])

  fig = go.Figure(grafico)
  fig.update_layout(title= 'Análise de cluster', xaxis_title= xlabel, yaxis_title= y_label)
  fig.show()
```

```python
plotar_graficos(base_completa['colesterol'], base_completa['peso'], 'colesterol', 'peso')
```
